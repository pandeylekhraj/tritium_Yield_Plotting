```python
import os
import numpy as np
import matplotlib.pyplot as plt

# File paths for Ge data
ge_file_paths = [
    "Run1_Ge_5e6_neg_muons_tritium_only_events.dat",
    "Run2_Ge_5e6_neg_muons_tritium_only_events.dat",
    "Run3_Ge_5e6_neg_muons_tritium_only_events.dat",
    "Run4_Ge_5e6_neg_muons_tritium_only_events.dat"
]

# File path for Si data
si_file_path = 'Si-tritium-KE-extract-FLUKA-107946_Lekhraj.dat'

# Function to read last column values from a file
def extract_last_column_values(file_paths):
    last_column_values = []
    for file_path in file_paths:
        if os.path.exists(file_path):
            with open(file_path, 'r') as file:
                for line in file:
                    parts = line.split()
                    try:
                        last_value = float(parts[-1])
                        # Multiply by 1000 to scale the value for plotting
                        last_column_values.append(last_value * 1000)
                    except (ValueError, IndexError):
                        print(f"Skipping line in {file_path}: {line.strip()}")
        else:
            print(f"File {file_path} not found.")
    return last_column_values

# Extracting last column values for Ge and Si data
ge_last_column_values = extract_last_column_values(ge_file_paths)
si_last_column_values = extract_last_column_values([si_file_path])

# Plotting the histograms for Ge and Si on the same plot with different colors
plt.figure(figsize=(8, 6))

# Ge histogram (using outline style in blue)
plt.hist(ge_last_column_values, bins=30, histtype='step', color='blue', linewidth=1.5, log=True, label='Tritium (in Ge)')

# Si histogram (using outline style in green)
plt.hist(si_last_column_values, bins=30, histtype='step', color='green', linewidth=1.5, log=True, label='Tritium (in Si)')

# Adding titles and labels
plt.xlabel("Kinetic Energy (MeV)")
plt.ylabel("Counts")
plt.legend()

# Display the combined histogram outline
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d, cKDTree

# Define the seed points for the Voronoi diagram
points = np.array([[0, 0], [0, 2], [1, 0], [1, 2],
                   [2, 0], [2, 1], [3, 3], [5, 5]])
data_points = np.random.randint(0, 6, size=(10, 2))  # Example random data points

# Create the Voronoi diagram
vor = Voronoi(points)

# Plot the Voronoi diagram
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='blue', line_width=1.5, point_size=0)
ax.scatter(data_points[:, 0], data_points[:, 1], color='red')
plt.title("Voronoi Diagram with Random Data Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()

# Initialize KDTree for observed count calculation
tree = cKDTree(points)
observed_counts = np.zeros(len(points), dtype=int)

# Calculate observed counts for each Voronoi cell
for point in data_points:
    dist, region_idx = tree.query(point)
    observed_counts[region_idx] += 1

# Calculate centroids for each Voronoi cell
centroids = []
for region_index in vor.point_region:
    region = vor.regions[region_index]
    
    # Skip infinite regions
    if -1 in region:
        centroids.append(None)  # None for infinite regions
        continue
    
    # Calculate centroid by averaging vertices
    vertices = vor.vertices[region]
    centroid = np.mean(vertices, axis=0)
    centroids.append(centroid)

# Display observed counts and centroids
for idx, (count, centroid) in enumerate(zip(observed_counts, centroids)):
    print(f"Voronoi cell {idx}: Observed Count = {count}")
    if centroid is not None:
        print(f" - Centroid at {centroid}")
    else:
        print(" - Centroid: Infinite region (No finite centroid)")

# Optional: Calculate integrated count based on distances to centroid
integrated_counts = np.zeros(len(points))
for idx, point in enumerate(data_points):
    dist, region_idx = tree.query(point)
    centroid = centroids[region_idx]
    print(centroid)
    if centroid is not None:  # Skip if centroid is infinite
        distance = np.linalg.norm(point - centroid)
        integrated_counts[region_idx] += 1 / (distance + 1e-9)  # Avoid division by zero
```
